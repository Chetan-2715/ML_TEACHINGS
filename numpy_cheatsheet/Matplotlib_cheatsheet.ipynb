```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.plot(x, y, label="cos(x)")
plt.plot(x, z, label="sin(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
plt.scatter(x, z, c="red", marker="o")
plt.title("Scatter Plot Example")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar([1, 2, 3], [3, 4, 5])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh([0.5, 1, 2.5], [0, 1, 2])
plt.show()
```

```python
plt.hist(y, bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram Example")
plt.show()
```

```python
plt.boxplot(y)
plt.title("Boxplot Example")
plt.show()
```

```python
plt.violinplot(z)
plt.title("Violin Plot Example")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='lightblue', linewidth=3)
ax.scatter([2, 4, 6], [5, 15, 25], color='darkgreen', marker='^')
ax.set_xlim(1, 6.5)
ax.set(title='Customized Axes', xlabel='X Axis', ylabel='Y Axis')
ax.legend(["cos(x)", "points"])
plt.show()
```

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap="viridis", interpolation="nearest")
plt.colorbar()
plt.title("Heatmap Example")
plt.show()
```

```python
plt.plot(x, y)
plt.text(5, 0, "Center Point", style="italic")
plt.annotate("Cosine Peak", xy=(0,1), xytext=(2,1.2),
             arrowprops=dict(arrowstyle="->"))
plt.title("Annotations Example")
plt.show()
```

```python
plt.plot(x, y, ls="--", lw=2, label="Dashed Line")
plt.plot(x, z, ls="-.", lw=2, label="Dash-dot Line")
plt.plot(x, x/10, ls=":", lw=2, label="Dotted Line")
plt.legend()
plt.title("Line Styles Example")
plt.show()
```

```python
plt.plot(x, y)
plt.show()

plt.cla()   # Clear an axis
plt.clf()   # Clear entire figure
plt.close() # Close window
```
