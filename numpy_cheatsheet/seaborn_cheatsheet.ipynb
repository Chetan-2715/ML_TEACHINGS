```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Built-in datasets
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")

iris.head()
```

```python
sns.set_style("whitegrid")
sns.set_context("talk", font_scale=1.2)
```

```python
sns.set_palette("husl", 3)
```

```python
sns.lmplot(x="sepal_width", y="sepal_length", data=iris)
plt.show()
```

```python
sns.regplot(x="tip", y="total_bill", data=tips)
plt.show()
```

```python
sns.stripplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
sns.barplot(x="sex", y="survived", hue="class", data=titanic)
plt.show()
```

```python
sns.countplot(x="deck", data=titanic, palette="Greens_d")
plt.show()
```

```python
sns.pointplot(x="class", y="survived", hue="sex", data=titanic,
              markers=["^","o"], linestyles=["-","--"])
plt.show()
```

```python
sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic)
plt.show()
```

```python
sns.violinplot(x="age", y="sex", hue="survived", data=titanic)
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True, color="b")
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.show()
```

```python
uniform_data = np.random.rand(10, 12)
sns.heatmap(uniform_data, vmin=0, vmax=1, cmap="coolwarm")
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="survived", row="sex")
g.map(plt.hist, "age")
plt.show()
```

```python
sns.lmplot(x="sepal_width", y="sepal_length", hue="species", data=iris)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x="sex", y="survived", data=titanic)
plt.title("Survival by Sex")
plt.xlabel("Passenger Sex")
plt.ylabel("Survival Rate")
plt.show()
```
